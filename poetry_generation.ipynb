```python
import pandas as pd
import numpy as np
import tensorflow as tf
import requests
import string
import pickle

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
data = response.text.splitlines()
len(data) #total snetences
```

```python
#We cant preprocess the data as we need the text as it is to generate new words in the poem

#TOKENIZATION

token = Tokenizer()
token.fit_on_texts(data)
token.word_counts
# pickle.dump(token,open("token.pkl","wb"))
```

```python
token.word_index #numerical rep of each word, numerical position
```

```python
encoded_text = token.texts_to_sequences(data)
encoded_text
```

```python
vocab_size = len(token.word_counts) + 1 #due to indexing. Given in Keras documentation
vocab_size
```

```python
#PREPARE TRAIN DATA

#Approach can be of any type, like keep a sliding window of size k = take first k words predict k+1, then take 2 to k+1 and predict k+2 etc
#we will take 1st word to predict 2nd , then 1 and 2 to get 3, 1+2+3=4,1+2+3+4=5 likewise

datalist = []
for sent in encoded_text:
  if len(sent)>1: #consider sentences with more than 1 word
    for i in range(2,len(sent)):
      datalist.append(sent[:i])
      # print(sent[:i])
```

```python
datalist[:10]
```

```python
#PADDING
max_len = 25

pad_text = pad_sequences(datalist, padding="pre", maxlen=max_len)
```

```python
pad_text
```

```python
pad_text.shape
```

```python
#we will add words one by one as already present in the datalist, and now all of them are of same length 25. Hence take 24 words from every item and the
# last one would be target

X = pad_text[:,:-1]
y = pad_text[:,-1]
y = to_categorical(y,num_classes=vocab_size)
seq_len = X.shape[1]
X.shape
```

```python
#LSTM MODEL BUILDING AND TRAINING

model = Sequential()

model.add(Embedding(vocab_size,50,input_length=seq_len)) #layer to generate the vector in NN.50size of NN

model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100))
model.add(Dense(100,activation="relu"))
model.add(Dense(vocab_size,activation="softmax"))

model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
model.fit(X,y,batch_size=32,epochs=100)  #X,y split up is not done as we are working with a single text.
```

```python
#POETRY GENERATION

poem_len = 10

def gen_poem(base_text,n_lines):
  for i in range(n_lines):
    final = []
    for _ in range(poem_len):
      enc = token.texts_to_sequences([base_text])
      enc = pad_sequences(enc,padding="pre",maxlen=seq_len) #here max length = 24 not 25 as we dont need to separate the target from input.
      pred = np.argmax(model.predict(enc),axis=-1) #will give the numeric value of that word
      
      pred_word =""
      for w,idx in token.word_index.items():
        if idx == pred:
          pred_word = w
          break
      base_text = base_text + " " + pred_word
      final.append(pred_word) #a sentence of length 10 will be formed.
    base_text = final[-1] #last word from previous line ll go as input to next
    final = " ".join(final)
    print(final) #this ll create n_lines
```

```python

# A simple Python3 program to find lexicographically
# minimum rotation of a given string
 
# This function return lexicographically minimum
# rotation of str
def minLexRotation(str_) :
 
    # Find length of given string
    n = len(str_)
 
    # Create an array of strings to store all rotations
    arr = [0] * n
 
    # Create a concatenation of string with itself
    concat = str_ + str_
 
    # One by one store all rotations of str in array.
    # A rotation is obtained by getting a substring of concat
    for i in range(n) :
        arr[i] = concat[i : n + i]
    x = arr.copy()
    # Sort all rotations
    arr.sort()
 
    # Return the first rotation from the sorted array
    return arr[0],arr,x
 
# Driver Code
# print(minLexRotation("GEEKSFORGEEKS"))
print(minLexRotation('decade'))#adcede
print(minLexRotation("bbbabbb"))#babbbbb
```

```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")


from random import randint
from collections import OrderedDict

class HashTable:
    def __init__(self, size):
        self.size = size
        self.data = [0] * size
        self.cache = OrderedDict()

    def __str__(self):
        return str(self.__dict__)

    def _hash(self, key):
        hash = 0
        for id, char in enumerate(key):
            hash = (hash + ord(char) * randint(0,10)) % len(self.data)
        return hash

    def add(self, key, value):
        adrs = self._hash(key)
        if self.data[adrs] != 0:
            self.data[adrs][0][1] = value
        else:
            self.data[adrs] = [[key, value]]
        self.cache.move_to_end(key)

    def get(self, key):
        adrs = self._hash(key)
        if self.data[adrs] == 0:
            value = -1
        else:
            count = len(self.data[adrs])
            for i in range(count):
                if self.data[adrs][i][0] == key:
                    value = self.data[adrs][i][1]
                    self.cache.move_to_end(key)
                    break
        return value

    def remove(self, key):
        adrs = self._hash(key)
        if self.data[adrs] == 0:
            value = -1
        else:
            count = len(self.data[adrs])
            for i in range(count):
                if self.data[adrs][i][0] == key:
                    value = self.data[adrs][i][1]
                    break
            self.data[adrs] = 0
        return value

    def evict(self):
        item = list(self.cache.keys())[0]
        rem = remove(item)


def operations(n):
    hasht = HashTable(len(n))
    out = []
    if not n:
        return []
    for ops in n:
        op = ops.split()
        if op[0] == "exit":
            break
        elif op[0] == "add":
            hasht.add(op[1],op[2])
        elif op[0] == "get" or op[0] == "remove":
            out.append(hasht.op[0](op[1]))
        else:
            hasht.evict()
    return out
```
